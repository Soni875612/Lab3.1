```python
# Sample dataset
df = pd.read_csv("India Agriculture Crop Production.csv")
df.head()  # Display first 5 rows of the dataset
```

```python
# Select Wheat and Rice data
wheat_df = df[df['Crop'] == 'Wheat']
rice_df = df[df['Crop'] == 'Rice']
```

```python
# Drop missing values in the Production column
wheat_df = wheat_df.dropna(subset=['Production'])
rice_df = rice_df.dropna(subset=['Production'])
```

```python
# Compute Mean, Median, and Mode for Wheat Production
mean_wheat = np.mean(wheat_df['Production'])
median_wheat = np.median(wheat_df['Production'])
mode_wheat = stats.mode(wheat_df['Production'], keepdims=True)[0][0]
```

```python
# Compute Mean, Median, and Mode for Rice Production
mean_rice = np.mean(rice_df['Production'])
median_rice = np.median(rice_df['Production'])
mode_rice = stats.mode(rice_df['Production'], keepdims=True)[0][0]
```

```python
# Print Results
print(f"Wheat Production - Mean: {mean_wheat}, Median: {median_wheat}, Mode: {mode_wheat}")
print(f"Rice Production - Mean: {mean_rice}, Median: {median_rice}, Mode: {mode_rice}")
```
